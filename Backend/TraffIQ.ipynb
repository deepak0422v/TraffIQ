```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv("jan to may police violation_anonymized791b166.csv")
df['created_datetime'] = pd.to_datetime(df['created_datetime'], format='ISO8601', utc=True)
df['date']        = df['created_datetime'].dt.date
df['hour']        = df['created_datetime'].dt.hour
df['week']        = df['created_datetime'].dt.isocalendar().week.astype(int)
df['is_peak']     = df['hour'].apply(lambda x: 1 if (7<=x<=10 or 17<=x<=21) else 0)
df['junction_clean'] = df['junction_name'].str.strip()
df['is_heavy']    = df['vehicle_type'].isin([
    'HGV','LORRY/GOODS VEHICLE','TANKER','BUS (BMTC/KSRTC)',
    'PRIVATE BUS','TOURIST BUS','SCHOOL VEHICLE'
]).astype(int)

daily = df.groupby(['junction_clean','date']).size().reset_index(name='daily_count')
daily['date'] = pd.to_datetime(daily['date'])
daily['week'] = daily['date'].dt.isocalendar().week.astype(int)

junction_features = daily.groupby('junction_clean').agg(
    daily_avg   = ('daily_count', 'mean'),
    daily_std   = ('daily_count', 'std'),
    daily_max   = ('daily_count', 'max'),
    daily_min   = ('daily_count', 'min'),
    total_days  = ('daily_count', 'count')
).fillna(0)

weekly = daily.groupby(['junction_clean','week'])['daily_count'].sum().reset_index()
first_4 = weekly.groupby('junction_clean').apply(
    lambda x: x.nsmallest(4,'week')['daily_count'].mean()
).rename('first_4w_avg')
last_4 = weekly.groupby('junction_clean').apply(
    lambda x: x.nlargest(4,'week')['daily_count'].mean()
).rename('last_4w_avg')

junction_features = junction_features.join(first_4).join(last_4)
junction_features['wow_growth'] = (
    (junction_features['last_4w_avg'] - junction_features['first_4w_avg']) /
    junction_features['first_4w_avg'].clip(lower=1)
)

junction_features['spike_ratio'] = (
    junction_features['daily_max'] / junction_features['daily_avg'].clip(lower=1)
)

junction_features['consistency'] = 1 - (
    junction_features['daily_std'] / junction_features['daily_avg'].clip(lower=1)
).clip(0, 1)

print("Features built:", junction_features.shape)
print(junction_features.describe().round(2))

features_for_cluster = ['daily_avg','consistency','wow_growth','spike_ratio']
X = junction_features[features_for_cluster].fillna(0)
X_scaled = StandardScaler().fit_transform(X)

kmeans = KMeans(n_clusters=4, random_state=42, n_init=20)
junction_features['cluster'] = kmeans.fit_predict(X_scaled)

cluster_profiles = junction_features.groupby('cluster')[features_for_cluster].mean()
print("\nCluster profiles:")
print(cluster_profiles.round(3))

labels = {}
for c in range(4):
    profile = cluster_profiles.loc[c]
    if profile['daily_avg'] > cluster_profiles['daily_avg'].median() and profile['consistency'] > 0.5:
        labels[c] = 'PERSISTENT'
    elif profile['spike_ratio'] > cluster_profiles['spike_ratio'].median() and profile['consistency'] < 0.4:
        labels[c] = 'BURST'
    elif profile['wow_growth'] > 0.1:
        labels[c] = 'EMERGING'
    else:
        labels[c] = 'STABLE'

junction_features['hotspot_type'] = junction_features['cluster'].map(labels)
print("\nHotspot types assigned:")
print(junction_features['hotspot_type'].value_counts())

print("\nTop junctions per type:")
for htype in ['PERSISTENT','BURST','EMERGING','STABLE']:
    subset = junction_features[junction_features['hotspot_type']==htype]
    if len(subset) > 0:
        top = subset.sort_values('daily_avg', ascending=False).head(3)
        print(f"\n{htype}:")
        print(top[['daily_avg','consistency','wow_growth','spike_ratio']].round(2).to_string())

junction_features.to_csv('junction_behavior.csv')
print("\njunction_behavior.csv saved")
```

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("jan to may police violation_anonymized791b166.csv")

# Parse datetime
df["created_datetime"] = pd.to_datetime(
    df["created_datetime"],
    format="ISO8601",
    utc=True
)

# Convert to IST
df["created_ist"] = (
    df["created_datetime"]
    .dt.tz_convert("Asia/Kolkata")
)

# Time features
df["date"] = df["created_ist"].dt.date
df["hour_ist"] = df["created_ist"].dt.hour
df["day_of_week"] = df["created_ist"].dt.dayofweek
df["month"] = df["created_ist"].dt.month

print(df.shape)
df.head()
```

```python
risk_data = (
    df[df["junction_name"] != "No Junction"]
    .groupby(
        [
            "junction_name",
            "police_station",
            "date",
            "hour_ist"
        ]
    )
    .size()
    .reset_index(name="violations")
)

risk_data["day_of_week"] = pd.to_datetime(
    risk_data["date"]
).dt.dayofweek

risk_data["month"] = pd.to_datetime(
    risk_data["date"]
).dt.month

risk_data["day"] = pd.to_datetime(
    risk_data["date"]
).dt.day

print(risk_data.shape)
risk_data.head()
```

```python
print(
    risk_data["junction_name"]
    .nunique()
)

print(
    risk_data["police_station"]
    .nunique()
)
```

```python
# Average violations per junction per hour

hourly_risk = (
    risk_data
    .groupby(
        ["junction_name", "hour_ist"]
    )["violations"]
    .mean()
    .reset_index()
)

hourly_risk.head()
```

```python
# Normalize risk inside each junction

hourly_risk["risk_score"] = (
    hourly_risk.groupby("junction_name")["violations"]
    .transform(
        lambda x: x / x.max()
    )
)

hourly_risk.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

junction_volume = (
    risk_data.groupby("junction_name")["violations"]
    .sum()
    .reset_index(name="total_violations")
)

scaler = MinMaxScaler()

junction_volume["volume_score"] = scaler.fit_transform(
    junction_volume[["total_violations"]]
)

junction_volume.head()
```

```python
hourly_risk_final = hourly_risk.merge(
    junction_volume,
    on="junction_name",
    how="left"
)

hourly_risk_final.head()
```

```python
hourly_risk_final["city_risk"] = (
    0.60 * hourly_risk_final["risk_score"]
    +
    0.40 * hourly_risk_final["volume_score"]
)

print(hourly_risk_final.columns)
```

```python
hourly_risk_final.sort_values(
    "city_risk",
    ascending=False
)[
    [
        "junction_name",
        "hour_ist",
        "violations",
        "risk_score",
        "volume_score",
        "city_risk"
    ]
].head(30)
```

```python
def recommend_deployment(current_hour, officers_available=20, top_n=10):

    current_risk = hourly_risk_final[
        hourly_risk_final["hour_ist"] == current_hour
    ].copy()

    current_risk = current_risk.sort_values(
        "city_risk",
        ascending=False
    )

    current_risk = current_risk.head(top_n)

    total_risk = current_risk["city_risk"].sum()

    current_risk["recommended_officers"] = (
        current_risk["city_risk"] / total_risk
        * officers_available
    ).round()

    current_risk["recommended_officers"] = (
        current_risk["recommended_officers"]
        .clip(lower=1)
        .astype(int)
    )

    return current_risk[
        [
            "junction_name",
            "city_risk",
            "recommended_officers"
        ]
    ]


deployment_10am = recommend_deployment(
    current_hour=10,
    officers_available=20
)

deployment_10am
```

```python
def calculate_expected_impact(row):

    risk = row["city_risk"]
    officers = row["recommended_officers"]

    violation_reduction = (
        risk * officers * 12
    )

    congestion_reduction = min(
        60,
        round(risk * officers * 8)
    )

    if risk >= 0.80:
        priority = "CRITICAL"

    elif risk >= 0.60:
        priority = "HIGH"

    elif risk >= 0.40:
        priority = "MEDIUM"

    else:
        priority = "LOW"

    return pd.Series([
        round(violation_reduction),
        congestion_reduction,
        priority
    ])


deployment_live = recommend_deployment(
    current_hour=10,
    officers_available=20
)

deployment_live[
    [
        "expected_violations_prevented",
        "expected_congestion_reduction",
        "priority"
    ]
] = deployment_live.apply(
    calculate_expected_impact,
    axis=1
)

deployment_live
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import numpy as np

# PDI REBUILD (WITHOUT NO JUNCTION)

junction_total = (
    named_df.groupby("junction_name")
    .size()
    .rename("total_violations")
)

junction_active_days = (
    named_df.groupby("junction_name")["date"]
    .nunique()
    .rename("active_days")
)

junction_peak = (
    named_df[named_df["hour_ist"].between(8,11)]
    .groupby("junction_name")
    .size()
    .rename("peak_violations")
)

heavy_types = [
    "HGV",
    "LORRY/GOODS VEHICLE",
    "TANKER",
    "PRIVATE BUS",
    "BUS (BMTC/KSRTC)",
    "TOURIST BUS",
    "SCHOOL VEHICLE"
]

junction_heavy = (
    named_df[named_df["vehicle_type"].isin(heavy_types)]
    .groupby("junction_name")
    .size()
    .rename("heavy_violations")
)

pdi = pd.concat([
    junction_total,
    junction_active_days,
    junction_peak,
    junction_heavy
], axis=1).fillna(0)

pdi["peak_ratio"] = (
    pdi["peak_violations"] /
    pdi["total_violations"]
)

pdi["heavy_ratio"] = (
    pdi["heavy_violations"] /
    pdi["total_violations"]
)

pdi["recurrence"] = (
    pdi["total_violations"] /
    pdi["active_days"]
)

scaler = MinMaxScaler()

pdi[[
    "freq_n",
    "peak_n",
    "heavy_n",
    "recur_n"
]] = scaler.fit_transform(
    pdi[[
        "total_violations",
        "peak_ratio",
        "heavy_ratio",
        "recurrence"
    ]]
)

pdi["PDI"] = (
      0.40 * pdi["freq_n"]
    + 0.25 * pdi["peak_n"]
    + 0.20 * pdi["heavy_n"]
    + 0.15 * pdi["recur_n"]
) * 100

pdi = pdi.sort_values(
    "PDI",
    ascending=False
)

print(pdi[[
    "total_violations",
    "peak_ratio",
    "heavy_ratio",
    "PDI"
]].head(20))
```

```python
import pandas as pd
import numpy as np

# Daily violations per junction

daily_counts = (
    named_df
    .groupby(["junction_name", "date"])
    .size()
    .reset_index(name="daily_violations")
)

daily_counts.head()
```

```python
junction_behavior = daily_counts.groupby(
    "junction_name"
).agg(
    daily_avg=("daily_violations", "mean"),
    daily_std=("daily_violations", "std"),
    daily_max=("daily_violations", "max"),
    active_days=("daily_violations", "count")
).reset_index()

junction_behavior["consistency"] = (
    1 -
    (
        junction_behavior["daily_std"]
        /
        junction_behavior["daily_avg"]
    )
).clip(0, 1)

junction_behavior["spike_ratio"] = (
    junction_behavior["daily_max"]
    /
    junction_behavior["daily_avg"]
)

junction_behavior = junction_behavior.fillna(0)

junction_behavior.head()
```

```python
daily_counts["week"] = (
    pd.to_datetime(daily_counts["date"])
    .dt.isocalendar()
    .week
)

weekly = (
    daily_counts
    .groupby(
        ["junction_name", "week"]
    )["daily_violations"]
    .sum()
    .reset_index()
)

growth_data = []

for junction in weekly["junction_name"].unique():

    temp = weekly[
        weekly["junction_name"] == junction
    ]

    if len(temp) < 8:
        growth = 0

    else:

        first4 = temp.head(4)["daily_violations"].mean()
        last4 = temp.tail(4)["daily_violations"].mean()

        growth = (
            (last4 - first4)
            /
            max(first4, 1)
        )

    growth_data.append(
        [junction, growth]
    )

growth_df = pd.DataFrame(
    growth_data,
    columns=[
        "junction_name",
        "growth_rate"
    ]
)

junction_behavior = junction_behavior.merge(
    growth_df,
    on="junction_name",
    how="left"
)

junction_behavior.head()
```

```python
behavior_df = junction_behavior.copy()

conditions = []

for _, row in behavior_df.iterrows():

    avg = row["daily_avg"]
    growth = row["growth_rate"]
    spike = row["spike_ratio"]

    if avg >= 40:
        conditions.append("PERSISTENT")

    elif growth >= 0.5:
        conditions.append("EMERGING")

    elif spike >= 6:
        conditions.append("BURST")

    else:
        conditions.append("STABLE")

behavior_df["hotspot_type"] = conditions

behavior_df["priority_level"] = np.where(
    behavior_df["daily_avg"] >= 50,
    "CRITICAL",
    np.where(
        behavior_df["daily_avg"] >= 20,
        "HIGH",
        np.where(
            behavior_df["daily_avg"] >= 8,
            "MEDIUM",
            "LOW"
        )
    )
)

behavior_df[
    [
        "junction_name",
        "daily_avg",
        "growth_rate",
        "spike_ratio",
        "hotspot_type",
        "priority_level"
    ]
].sort_values(
    "daily_avg",
    ascending=False
).head(30)
```

```python
# Rebuild final PDI using named_df

junction_total = (
    named_df.groupby("junction_name")
    .size()
    .rename("total_violations")
)

junction_peak = (
    named_df[named_df["hour_ist"].between(7,10)]
    .groupby("junction_name")
    .size()
    .rename("peak_violations")
)

junction_heavy = (
    named_df[
        named_df["vehicle_type"].isin([
            "HGV",
            "LORRY/GOODS VEHICLE",
            "TANKER",
            "BUS (BMTC/KSRTC)",
            "PRIVATE BUS",
            "TOURIST BUS",
            "SCHOOL VEHICLE"
        ])
    ]
    .groupby("junction_name")
    .size()
    .rename("heavy_violations")
)

active_days = (
    named_df.groupby("junction_name")["date"]
    .nunique()
    .rename("active_days")
)

junction_pdi_final = pd.concat(
    [
        junction_total,
        active_days,
        junction_peak,
        junction_heavy
    ],
    axis=1
).fillna(0)

junction_pdi_final["peak_ratio"] = (
    junction_pdi_final["peak_violations"]
    /
    junction_pdi_final["total_violations"]
)

junction_pdi_final["heavy_ratio"] = (
    junction_pdi_final["heavy_violations"]
    /
    junction_pdi_final["total_violations"]
)

junction_pdi_final["recurrence"] = (
    junction_pdi_final["total_violations"]
    /
    junction_pdi_final["active_days"]
)

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

junction_pdi_final["freq_n"] = scaler.fit_transform(
    junction_pdi_final[["total_violations"]]
)

junction_pdi_final["peak_n"] = scaler.fit_transform(
    junction_pdi_final[["peak_ratio"]]
)

junction_pdi_final["heavy_n"] = scaler.fit_transform(
    junction_pdi_final[["heavy_ratio"]]
)

junction_pdi_final["recur_n"] = scaler.fit_transform(
    junction_pdi_final[["recurrence"]]
)

junction_pdi_final["PDI"] = (
    0.45 * junction_pdi_final["freq_n"] +
    0.25 * junction_pdi_final["peak_n"] +
    0.10 * junction_pdi_final["heavy_n"] +
    0.20 * junction_pdi_final["recur_n"]
) * 100

junction_pdi_final.to_csv(
    "junction_pdi_final.csv"
)

print(junction_pdi_final.shape)

junction_pdi_final.sort_values(
    "PDI",
    ascending=False
).head(20)
```
